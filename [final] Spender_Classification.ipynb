```python
dummies1=pd.get_dummies(df2['ADJATTRNETWORK'])
dummies1=dummies1.drop(dummies1.columns[0],axis='columns')

dummies2=pd.get_dummies(df2['PLATFORM'])
dummies2=dummies2.drop(dummies2.columns[0],axis='columns')

dummies3=pd.get_dummies(df2['COUNTRY'])
dummies3=dummies3.drop(dummies3.columns[0],axis='columns')
```

```python
df2=df2.drop(['COUNTRY','PLATFORM','ADJATTRNETWORK'],axis='columns')
df2=pd.concat([df2,dummies1,dummies2,dummies3],axis='columns')
```

```python
Y=df2['LTV_spender']
X=df2.drop(['LTV_spender'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=55)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(objective='count:poisson',n_estimators=250)
xgb.fit(X,Y)
```

```python
xgb.score(X_test,Y_test)
```

```python
Y_predicted=xgb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predicted)
sns.heatmap(cm,annot=True,fmt='1.0f')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_predicted))
```

```python
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(Y_test,Y_predicted)
auc=metrics.roc_auc_score(Y_test,Y_predicted)
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
sns.set()
i = np.arange(len(tpr)) # index for df
roc = pd.DataFrame({'fpr' : pd.Series(fpr, index=i),'tpr' : pd.Series(tpr, index = i), '1-fpr' : pd.Series(1-fpr, index = i), 'tf' : pd.Series(tpr - (1-fpr), index = i), 'thresholds' : pd.Series(thresholds, index = i)})
roc.iloc[(roc.tf-0).abs().argsort()[:1]]

# Plot tpr vs 1-fpr
fig, ax = plt.subplots()
plt.plot(roc['tpr'])
plt.plot(roc['1-fpr'], color = 'red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
ax.set_xticklabels([])
plt.show()
```
